```python
df.replace("", np.nan, inplace=True)
df.replace(" ", np.nan, inplace=True)
df.replace("None", np.nan, inplace=True)
```

```python
df["CustomerName"]=df["CustomerName"].str.title()

df["Gender"] = df["Gender"].str.capitalize()
df["PaymentMode"] = df["PaymentMode"].str.title()

df["Salesperson"] = df["Salesperson"].str.title()

df["Returned"] = df["Returned"].str.capitalize()
```

```python
df["Region"] = df["Region"].replace({
    "Nroth":"North",
    "south":"South",
    "south ":"South"
})
```

```python
df["Gender"] = df["Gender"].replace({
    "Male":"Male",
    "male":"Male",
    "Female":"Female",
    "F":"Female"
})
```

```python
df["PaymentMode"] = df["PaymentMode"].replace({
    "card":"Card"
})
```

```python
df["Returned"] = df["Returned"].replace({
    "NO":"No",
    "YES":"Yes"
})
```

```python
df["Age"] = df["Age"].replace({
    "Twenty Eight":28
})

df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors='coerce')
```

```python
df["UnitPrice"] = pd.to_numeric(df["UnitPrice"], errors='coerce')
```

```python
df["Discount"] = (
    df["Discount"]
    .astype(str)
    .str.replace("%","", regex=False)
    .str.strip()
)
```

```python
#1. Sales by Region (Bar Chart)

import matplotlib.pyplot as plt

sales_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sales_region.plot(kind="bar")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()



#Insight
#This chart shows which region contributes the most revenue.
#Regions with lower sales may require additional marketing efforts.
```

```python
#2. Product-wise Sales (Bar Chart)
product_sales = df.groupby("Product")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
product_sales.plot(kind="bar")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()

#Insight
#Identifies the best-selling products.
#High-performing products should be prioritized for inventory and promotions.
```

```python
#3. Payment Mode Distribution (Pie Chart)
payment = df["PaymentMode"].value_counts()

plt.figure(figsize=(7,7))
payment.plot(
    kind="pie",
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Payment Mode Distribution")
plt.ylabel("")
plt.show()

#Insight
#Displays customer payment preferences.
#Businesses can focus on the most frequently used payment methods.
```

```python
#4. Sales Distribution (Histogram)
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=20)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()

#Insight
#Illustrates how sales values are distributed.
#A few very high sales may indicate premium or bulk purchases.
```

```python
#5. Category-wise Sales
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(6,5))

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.xticks(rotation=0)

plt.show()

#Insight
#Compares sales across product categories.
#Helps determine which category contributes the most revenue.
```
